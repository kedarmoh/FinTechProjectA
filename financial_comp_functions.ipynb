```python
import json
```

```python
def ticker_financial_ratio(stock):
    """
    Enter a stock ticker to get the company's financial ratio. 
    """
    ticker = str(stock)
    response = urlopen(
        'https://financialmodelingprep.com/api/v3/ratios/' + ticker + '?apikey=' + fmp_api_key
    )
    data = response.read().decode('utf-8')
    return json.loads(data)
```

```python
def tickers_financial_ratio(*args):
    tickers = [arg for arg in args]
    data = {}
    for ticker in tickers: 
        response = urlopen(
        'https://financialmodelingprep.com/api/v3/ratios/' + ticker + '?apikey=' + fmp_api_key
        )
        ticker_data = response.read().decode('utf-8')
        data[ticker] = json.loads(ticker_data)
    return data
        
```
